## Solution 

```python
%%writefile edge_pressure.F90

subroutine compute_edge_pressure(is, ie, js, je, isd, ied, jsd, jed, npz, ptop, pe, delp)
    integer, intent(in) :: is, ie, js, je, isd, ied, jsd, jed, npz
    double precision, intent(in) :: ptop
    double precision, intent(in), dimension(isd:ied,jsd:jed,npz):: delp
    double precision, intent(inout), dimension(is-1:ie+1,npz+1,js-1:je+1):: pe
    integer:: i,j,k

    do j=js,je
     pe(is-1,1,j) = ptop
     pe(ie+1,1,j) = ptop
     do k=1,npz
        pe(is-1,k+1,j) = pe(is-1,k,j) + delp(is-1,j,k)
        pe(ie+1,k+1,j) = pe(ie+1,k,j) + delp(ie+1,j,k)
     enddo
    enddo
    
    do i=is-1,ie+1
     pe(i,1,js-1) = ptop
     pe(i,1,je+1) = ptop
     do k=1,npz
        pe(i,k+1,js-1) = pe(i,k,js-1) + delp(i,js-1,k)
        pe(i,k+1,je+1) = pe(i,k,je+1) + delp(i,je+1,k)
     enddo
  enddo
end subroutine

program edge_pressure

  implicit none
    
  integer :: is, ie, js, je, isd, ied, jsd, jed, npz
  double precision :: ptop
  double precision, allocatable, dimension(:,:,:) :: delp, pe
  !$ser init directory='./data_pe' prefix='port' unique_id=.true.
  !$ser mode write
  !$ser on
  is = 1
  ie = 48
  js = 1
  je = 48
  isd = -2
  ied = 51
  jsd = -2
  jed = 51
  npz = 79
  ptop = 300.0
  allocate(delp(isd:ied, jsd:jed, npz))
  allocate(pe(is-1:ie+1,npz+1,js-1:je+1))
  !$ser savepoint 'edge_pressure-in'
  !$ser data ptop=ptop pe=pe delp=delp nhalo=is-isd nx=ie-is+1 ny=je-js+1 nz=npz
  call compute_edge_pressure(is, ie, js, je, isd, ied, jsd, jed, npz, ptop, pe, delp)
  !$ser savepoint 'edge_pressure-out'
  !$ser data pe=pe
  !$ser cleanup
end program
```

### Python Solution A
Using array size that is the max shape of all data arrays, matching indices. Included here are both solutions using regions and not using regions. 

```python
#!/usr/bin/env python3

import numpy as np
import gt4py.gtscript as gtscript
import gt4py.storage as gt_storage
import sys
import os
sys.path.append(os.environ.get('SERIALBOX_ROOT')+ '/python')
import serialbox as ser

field = gtscript.Field[np.float64]
backend="numpy"
origin=(0, 0, 0)

@gtscript.stencil(backend=backend)
def edge_pressure(pe: field, delp: field, ptop: float):
    with computation(FORWARD):
        with interval(0, 1):
            pe = ptop
        with interval(1, None):
            pe = pe[0, 0, -1] + delp[0, 0, -1]
            

def compute(pe, delp, ptop, nhalo, nx, ny, nz):
    istart = nhalo - 1
    edge_domain_x = (1, ny, nz + 1)
    edge_pressure(pe, delp, ptop, origin=(istart, nhalo, 0), domain=edge_domain_x)
    edge_pressure(pe, delp, ptop, origin=(nhalo + nx, nhalo, 0), domain=edge_domain_x)
    edge_domain_y = (nx + 2, 1, nz + 1)
    edge_pressure(pe, delp, ptop, origin=(istart, nhalo - 1, 0), domain=edge_domain_y)
    edge_pressure(pe, delp, ptop, origin=(istart, nhalo + ny , 0), domain=edge_domain_y)

    
@gtscript.stencil(backend=backend)
def edge_pressure_regions(pe: field, delp: field, ptop: float):
    with computation(FORWARD):
        with interval(0, 1):
            with parallel(region[I[0], :], region[I[-1], :], region[:, J[0]], region[:, J[-1]]):
                pe = ptop
            
        with interval(1, None):
            with parallel(region[I[0], :], region[I[-1], :], region[:, J[0]], region[:, J[-1]]):
                pe = pe[0, 0, -1] + delp[0, 0, -1]
 

def serialized_scalar(serializer, savepoint, name):
    return serializer.read(name,  savepoint[0])[0]

serializer = ser.Serializer(ser.OpenModeKind.Read,'./data_pe', 'port')

sp_in = serializer.get_savepoint('edge_pressure-in')

pe = serializer.read('pe',  sp_in[0])
delp = serializer.read('delp',  sp_in[0])
ptop = serializer.read('ptop',  sp_in[0])[0]
nhalo = serialized_scalar(serializer, sp_in, 'nhalo')
nx = serialized_scalar(serializer, sp_in, 'nx')
ny = serialized_scalar(serializer, sp_in, 'ny')
nz = serialized_scalar(serializer, sp_in, 'nz')

# reshape pe it be i, j, k
pe = np.moveaxis(pe, 1, 2)

# pick a storage size that encompasses both delp and pe
storage_shape = (nx + 2 * nhalo, ny + 2 * nhalo, nz + 1)

# make pe gt storage
pe_overlap = np.empty(storage_shape)
pe_overlap[nhalo - 1:nhalo + nx + 1, nhalo - 1: nhalo + ny + 1, :] = pe
pe = gt_storage.from_array(data=pe_overlap, backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape)
pe_regions = gt_storage.from_array(data=pe_overlap.data, backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape)
# make delp storage
delp_overlap = np.empty(storage_shape)
delp_overlap[:, :, 0:nz] = delp
delp = gt_storage.from_array(data=delp_overlap,  backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape,)
# call the computation
compute(pe, delp, ptop, nhalo, nx, ny, nz)
# compute using regions instead
edge_pressure_regions(pe_regions, delp, ptop, origin=(nhalo - 1, nhalo - 1, 0), domain=(nx + 2, ny + 2, nz + 1))

# Get regression data
sp_out = serializer.get_savepoint('edge_pressure-out')
pe_ref = serializer.read('pe',  sp_out[0])

# reformat computed value back
pe = pe[nhalo - 1:nhalo + nx + 1, nhalo - 1: nhalo + ny + 1, :] 
pe = np.moveaxis(pe, 2, 1)
pe_regions = pe_regions[nhalo - 1:nhalo + nx + 1, nhalo - 1: nhalo + ny + 1, :] 
pe_regions = np.moveaxis(pe_regions, 2, 1)
# compare answers
try:
    assert np.array_equal(pe_ref, pe), "pe does not match!"
    assert np.array_equal(pe_ref, pe_regions), "pe_regions does not match!"
except AssertionError as msg:
    print(msg)
    
print("Finished running comparison tests!")
```

## Solution B
Using array size that is the minimum size required for this computation. Included here are both solutions using regions and not using regions. 

```python
#!/usr/bin/env python3

import numpy as np
import gt4py.gtscript as gtscript
import gt4py.storage as gt_storage
import sys
import os
sys.path.append(os.environ.get('SERIALBOX_ROOT')+ '/python')
import serialbox as ser

field = gtscript.Field[np.float64]
backend="numpy"
origin=(0, 0, 0)
@gtscript.stencil(backend=backend)
def edge_pressure(pe: field, delp: field, ptop: float):
    with computation(FORWARD):
        with interval(0, 1):
            pe = ptop
        with interval(1, None):
            pe = pe[0, 0, -1] + delp[0, 0, -1]

def compute(pe, delp, ptop, nx, ny, nz):
    edge_domain_x = (1, ny, nz + 1)
    edge_pressure(pe, delp, ptop, origin=(0, 1, 0), domain=edge_domain_x)
    edge_pressure(pe, delp, ptop, origin=(nx + 1, 1, 0), domain=edge_domain_x)
    edge_domain_y = (nx + 2, 1, nz + 1)
    edge_pressure(pe, delp, ptop, origin=(0, 0, 0), domain=edge_domain_y)
    edge_pressure(pe, delp, ptop, origin=(0, ny + 1, 0), domain=edge_domain_y)

@gtscript.stencil(backend=backend)
def edge_pressure_regions(pe: field, delp: field, ptop: float):
    with computation(FORWARD):
        with interval(0, 1):
            with parallel(region[I[0], :], region[I[-1], :], region[:, J[0]], region[:, J[-1]]):
                pe = ptop
            
        with interval(1, None):
            with parallel(region[I[0], :], region[I[-1], :], region[:, J[0]], region[:, J[-1]]):
                pe = pe[0, 0, -1] + delp[0, 0, -1]
 
def serialized_scalar(serializer, savepoint, name):
    return serializer.read(name,  savepoint[0])[0]
serializer = ser.Serializer(ser.OpenModeKind.Read,'./data_pe', 'port')

sp_in = serializer.get_savepoint('edge_pressure-in')

pe = serializer.read('pe',  sp_in[0])
delp = serializer.read('delp',  sp_in[0])
ptop = serializer.read('ptop',  sp_in[0])[0]
nhalo = serialized_scalar(serializer, sp_in, 'nhalo')
nx = serialized_scalar(serializer, sp_in, 'nx')
ny = serialized_scalar(serializer, sp_in, 'ny')
nz = serialized_scalar(serializer, sp_in, 'nz')

# reshape pe it be i, j, k
pe = np.moveaxis(pe, 1, 2)

# Use the smallest storage size necessary
storage_shape=pe.shape

# make pe gt storage
pe =  gt_storage.from_array(data=pe, backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape)
pe_regions = gt_storage.from_array(data=pe.data, backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape)

# make delp storage, only the subset used in this computation
delp_overlap = np.empty(storage_shape)
delp_overlap[:, :, 0:nz] = delp[nhalo - 1: nhalo + nx + 1, nhalo - 1: nhalo + ny + 1,:]
delp = gt_storage.from_array(data=delp_overlap,  backend=backend,  dtype=np.float64,
                            default_origin=origin, shape=storage_shape,)
# compute
compute(pe, delp, ptop, nx, ny, nz)
# compute using regions instead
edge_pressure_regions(pe_regions, delp, ptop, origin=(0, 0, 0), domain=(nx + 2, ny + 2, nz + 1))

# Get regression data
sp_out = serializer.get_savepoint('edge_pressure-out')
pe_ref = serializer.read('pe',  sp_out[0])

# reformat computed value back 
pe =  np.moveaxis(pe, 2, 1)
pe_regions = np.moveaxis(pe_regions, 2, 1)

# compare answers
try:
    assert np.array_equal(pe_ref, pe), "pe does not match!"
    assert np.array_equal(pe_ref, pe_regions), "pe_regions does not match!"
except AssertionError as msg:
    print(msg)
    
print("Finished running comparison tests!")
```
